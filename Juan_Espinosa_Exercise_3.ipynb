### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv).

### Step 3. Assign it to a variable called drinks.

```python
drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks.head()
```

### Step 4. Which continent drinks more beer on average?

```python
drinks_beer = drinks.groupby('continent').beer_servings.mean()    # Agrupar la cantidad de cerveza bebida por continente
drinks_beer = drinks_beer.sort_values(ascending=False)            # Ordenar de mayor a menor
drinks_beer.head(1)                                              # Imprimir el primer valor
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks_wine = drinks.groupby('continent').wine_servings.describe()
drinks_wine
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks_cont = drinks.groupby('continent').mean(numeric_only = True)
drinks_cont
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
drinks_cont2 = drinks.groupby('continent').median(numeric_only = True)
drinks_cont2
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
drinks_spirit = drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
drinks_spirit
```
