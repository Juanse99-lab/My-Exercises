### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep="\t")
```

### Step 4. How many products cost more than $10.00?

```python
# Eliminar signo $ de item_price
chipo['item_price'] = chipo['item_price'].str.replace('$', '')
# Convertir columna item_price a float
lambda x: float(x[0:-1])
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[0:-1]))
```

```python
chipo_10 = chipo['item_price'] > 10     # Filtrar precios superiores a $10
chipo_10 = chipo[chipo_10]              # Aplicar filtro
print(f'Hay {chipo_10.shape[0]} productos cuyo precio es superior a $10')   # Conteo de productos con precio > $10
chipo_10.head()
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
chipo_each = chipo['quantity'] == 1     # Filtrar cantidades = 1
chipo_each = chipo[chipo_each]           # Aplicar filtro
chipo_each.drop_duplicates(['item_name', 'choice_description'])     # Eliminar duplicados
chipo_each[['item_name','item_price']]     # Mostrar columnas item_name y item_price
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values('item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo_max_price = chipo.sort_values('item_price', ascending=False)      # Ordenar productos de mayor a menor precio
print(f'Se ordenaron {chipo_max_price["quantity"][3598]} unidades del producto más caro')     # Mostrar resultado
chipo_max_price.head(1)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo_veggie = chipo['item_name'] == 'Veggie Salad Bowl'     # Filtrar Veggie Salad Bowl
chipo_veggie = chipo[chipo_veggie]                           # Aplicar filtro
print(f'Se ordenaron {chipo_veggie["order_id"].count()} unidades de Veggie Salad Bowl')  # Conteo de productos Veggie Salad Bowl
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo_soda = chipo['item_name'] == 'Canned Soda'     # Filtrar Canned Soda
chipo_soda = chipo[chipo_soda]                       # Aplicar filtro
chipo_soda_2 = chipo_soda['quantity'] > 1            # Filtrar cantidades > 1
chipo_soda_2 = chipo_soda[chipo_soda_2]              # Aplicar filtro
chipo_soda_2.shape[0]                                # Conteo de productos Canned Soda > 1
```
